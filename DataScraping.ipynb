```python
def zip_file():
    year_file_list=[]
    for i in range(2005,2018):
        year_file_list.append('sample_'+str(i)+'.zip')
    return year_file_list
```

```python
def download_zip():
    year_file_list=zip_file()
    for j in year_file_list:
        driver.click(j)
        time.sleep(60)
```

```python
def extract_zip():
    year_file_list=zip_file()
    current_path=os.getcwd()
    path=current_path+'\ExtractData'
    download_path=current_path.split('Desktop')[0]+'Downloads\\'
    for i in year_file_list:
        with ZipFile(download_path+i, 'r') as zip:
            zip.extractall(path)
```

```python
def fill_null_values_orig(orig_dataset):
    orig_dataset['CREDIT SCORE']=orig_dataset['CREDIT SCORE'].fillna(value=9999)
    orig_dataset['FIRST PAYMENT DATE']=orig_dataset['FIRST PAYMENT DATE'].fillna(value='189901')
    orig_dataset['FIRST TIME HOMEBUYER FLAG']=orig_dataset['FIRST TIME HOMEBUYER FLAG'].fillna(value=9)
    orig_dataset['MATURITY DATE']=orig_dataset['MATURITY DATE'].fillna(value='189901')
    orig_dataset['METROPOLITAN STATISTICAL AREA (MSA)']=orig_dataset['METROPOLITAN STATISTICAL AREA (MSA)'].fillna(value=0)
    orig_dataset['MORTGAGE INSURANCE PERCENTAGE (MI %)']=orig_dataset['MORTGAGE INSURANCE PERCENTAGE (MI %)'].fillna(value=999)
    orig_dataset['NUMBER OF UNITS']=orig_dataset['NUMBER OF UNITS'].fillna(value=99)
    orig_dataset['OCCUPANCY STATUS']=orig_dataset['OCCUPANCY STATUS'].fillna(value=9)
    orig_dataset['ORIGINAL COMBINED LOAN-TO-VALUE (CLTV)']=orig_dataset['ORIGINAL COMBINED LOAN-TO-VALUE (CLTV)'].fillna(value=999)
    orig_dataset['ORIGINAL DEBT-TO-INCOME (DTI) RATIO']=orig_dataset['ORIGINAL DEBT-TO-INCOME (DTI) RATIO'].fillna(value=999)
    orig_dataset['ORIGINAL UPB']=orig_dataset['ORIGINAL UPB'].fillna(value=0)
    orig_dataset['ORIGINAL LOAN-TO-VALUE (LTV)']=orig_dataset['ORIGINAL LOAN-TO-VALUE (LTV)'].fillna(value=999)
    orig_dataset['ORIGINAL INTEREST RATE']=orig_dataset['ORIGINAL INTEREST RATE'].fillna(value=-1)
    orig_dataset['CHANNEL']=orig_dataset['CHANNEL'].fillna(value=9)
    orig_dataset['PREPAYMENT PENALTY MORTGAGE (PPM) FLAG']=orig_dataset['PREPAYMENT PENALTY MORTGAGE (PPM) FLAG'].fillna(value='X')
    orig_dataset['PRODUCT TYPE']=orig_dataset['PRODUCT TYPE'].fillna(value='NA')
    orig_dataset['PROPERTY STATE']=orig_dataset['PROPERTY STATE'].fillna(value='XX')
    orig_dataset['PROPERTY TYPE']=orig_dataset['PROPERTY TYPE'].fillna(value=99)
    orig_dataset['POSTAL CODE']=orig_dataset['POSTAL CODE'].fillna(value=0)
    orig_dataset['LOAN SEQUENCE NUMBER']=orig_dataset['LOAN SEQUENCE NUMBER'].fillna(value='X')
    orig_dataset['LOAN PURPOSE']=orig_dataset['LOAN PURPOSE'].fillna(value=9)
    orig_dataset['ORIGINAL LOAN TERM']=orig_dataset['ORIGINAL LOAN TERM'].fillna(value=-1)
    orig_dataset['NUMBER OF BORROWERS']=orig_dataset['NUMBER OF BORROWERS'].fillna(value=99)
    orig_dataset['SELLER NAME']=orig_dataset['SELLER NAME'].fillna(value='Other Sellers')
    orig_dataset['SERVICER NAME']=orig_dataset['SERVICER NAME'].fillna(value='Other Servicers')
    orig_dataset['SUPER CONFORMING FLAG']=orig_dataset['SUPER CONFORMING FLAG'].fillna(value='X')
    orig_dataset['Pre-HARP LOAN SEQUENCE NUMBER']=orig_dataset['Pre-HARP LOAN SEQUENCE NUMBER'].fillna(value='X')
    return orig_dataset
```

```python
def fill_null_values_svcg(svcg_dataset):
    svcg_dataset['LOAN SEQUENCE NUMBER']=svcg_dataset['LOAN SEQUENCE NUMBER'].fillna(value='X')
    svcg_dataset['MONTHLY REPORTING PERIOD']=svcg_dataset['MONTHLY REPORTING PERIOD'].fillna(value='189901')
    svcg_dataset['CURRENT ACTUAL UPB']=svcg_dataset['CURRENT ACTUAL UPB'].fillna(value=0)
    svcg_dataset['CURRENT LOAN DELINQUENCY STATUS']=svcg_dataset['CURRENT LOAN DELINQUENCY STATUS'].fillna(value='XX')
    svcg_dataset['LOAN AGE']=svcg_dataset['LOAN AGE'].fillna(value=-1)
    svcg_dataset['REMAINING MONTHS TO LEGAL MATURITY']=svcg_dataset['REMAINING MONTHS TO LEGAL MATURITY'].fillna(value=-1)
    svcg_dataset['REPURCHASE FLAG']=svcg_dataset['REPURCHASE FLAG'].fillna(value='X')
    svcg_dataset['MODIFICATION FLAG']=svcg_dataset['MODIFICATION FLAG'].fillna(value='X')
    svcg_dataset['ZERO BALANCE CODE']=svcg_dataset['ZERO BALANCE CODE'].fillna(value=-1)
    svcg_dataset['ZERO BALANCE EFFECTIVE DATE']=svcg_dataset['ZERO BALANCE EFFECTIVE DATE'].fillna(value='189901')
    svcg_dataset['CURRENT INTEREST RATE']=svcg_dataset['CURRENT INTEREST RATE'].fillna(value=-1)
    svcg_dataset['CURRENT DEFERRED UPB']=svcg_dataset['CURRENT DEFERRED UPB'].fillna(value=0)
    svcg_dataset['DUE DATE OF LAST PAID INSTALLMENT (DDLPI)']=svcg_dataset['DUE DATE OF LAST PAID INSTALLMENT (DDLPI)'].fillna(value='189901')
    svcg_dataset['MI RECOVERIES']=svcg_dataset['MI RECOVERIES'].fillna(value=0)
    svcg_dataset['NET SALES PROCEEDS']=svcg_dataset['NET SALES PROCEEDS'].fillna(value='U')
    svcg_dataset['NON MI RECOVERIES']=svcg_dataset['NON MI RECOVERIES'].fillna(value=0)
    svcg_dataset['EXPENSES']=svcg_dataset['EXPENSES'].fillna(value=-1)
    svcg_dataset['LEGAL COSTS']=svcg_dataset['LEGAL COSTS'].fillna(value=-1)
    svcg_dataset['MAINTENANCE AND PRESERVATION COSTS']=svcg_dataset['MAINTENANCE AND PRESERVATION COSTS'].fillna(value=-1)
    svcg_dataset['TAXES AND INSURANCE']=svcg_dataset['TAXES AND INSURANCE'].fillna(value=0)
    svcg_dataset['MISCELLANEOUS EXPENSES']=svcg_dataset['MISCELLANEOUS EXPENSES'].fillna(value=-1)
    svcg_dataset['ACTUAL LOSS CALCULATION']=svcg_dataset['ACTUAL LOSS CALCULATION'].fillna(value=0)
    svcg_dataset['MODIFICATION COST']=svcg_dataset['MODIFICATION COST'].fillna(value=-1)
    svcg_dataset['STEP MODIFICATION FLAG']=svcg_dataset['STEP MODIFICATION FLAG'].fillna(value='X')
    svcg_dataset['DEFERRED PAYMENT MODIFICATION']=svcg_dataset['DEFERRED PAYMENT MODIFICATION'].fillna(value='X')
    svcg_dataset['ESTIMATED LOAN TO VALUE (ELTV)']=svcg_dataset['ESTIMATED LOAN TO VALUE (ELTV)'].fillna(value=-1)
    return svcg_dataset
```

```python
def orig_dataframe():
    current_path=os.getcwd()
    path=current_path+'\ExtractData'
    orig_column_names=['CREDIT SCORE','FIRST PAYMENT DATE','FIRST TIME HOMEBUYER FLAG','MATURITY DATE',
              'METROPOLITAN STATISTICAL AREA (MSA)','MORTGAGE INSURANCE PERCENTAGE (MI %)','NUMBER OF UNITS',
              'OCCUPANCY STATUS','ORIGINAL COMBINED LOAN-TO-VALUE (CLTV)','ORIGINAL DEBT-TO-INCOME (DTI) RATIO','ORIGINAL UPB',
              'ORIGINAL LOAN-TO-VALUE (LTV)','ORIGINAL INTEREST RATE','CHANNEL','PREPAYMENT PENALTY MORTGAGE (PPM) FLAG',
              'PRODUCT TYPE','PROPERTY STATE','PROPERTY TYPE','POSTAL CODE','LOAN SEQUENCE NUMBER','LOAN PURPOSE',
              'ORIGINAL LOAN TERM','NUMBER OF BORROWERS','SELLER NAME','SERVICER NAME','SUPER CONFORMING FLAG',
              'Pre-HARP LOAN SEQUENCE NUMBER']
    
    orig_dataset=pd.DataFrame()
    for filename in os.listdir(path):
        if 'orig' in filename:
            print(filename)
            orig_dataset=orig_dataset.append(pd.read_csv(path+'\\'+filename,sep='|',names=orig_column_names,low_memory=False))
    return orig_dataset
```

```python
def change_datatype_orig(orig_dataset):
    
    original_dataset[['METROPOLITAN STATISTICAL AREA (MSA)','POSTAL CODE']] = original_dataset[['METROPOLITAN STATISTICAL AREA (MSA)','POSTAL CODE']].astype('int64')
    
    original_dataset[['Pre-HARP LOAN SEQUENCE NUMBER']] = original_dataset[['Pre-HARP LOAN SEQUENCE NUMBER']].astype('str')
    return original_dataset
```

```python
def svcg_dataframe():
    current_path=os.getcwd()
    path=current_path+'\ExtractData'
    svcg_column_names=['LOAN SEQUENCE NUMBER','MONTHLY REPORTING PERIOD','CURRENT ACTUAL UPB','CURRENT LOAN DELINQUENCY STATUS',
                  'LOAN AGE','REMAINING MONTHS TO LEGAL MATURITY','REPURCHASE FLAG','MODIFICATION FLAG','ZERO BALANCE CODE',
                  'ZERO BALANCE EFFECTIVE DATE','CURRENT INTEREST RATE','CURRENT DEFERRED UPB',
                   'DUE DATE OF LAST PAID INSTALLMENT (DDLPI)','MI RECOVERIES','NET SALES PROCEEDS','NON MI RECOVERIES',
                  'EXPENSES','LEGAL COSTS','MAINTENANCE AND PRESERVATION COSTS','TAXES AND INSURANCE','MISCELLANEOUS EXPENSES',
                  'ACTUAL LOSS CALCULATION','MODIFICATION COST','STEP MODIFICATION FLAG','DEFERRED PAYMENT MODIFICATION',
                  'ESTIMATED LOAN TO VALUE (ELTV)']
    svcg_dataset=pd.DataFrame()
    for filename in os.listdir(path):
        if 'svcg' in filename:
            print(filename)
            svcg_dataset=svcg_dataset.append(pd.read_csv(path+'\\'+filename,sep='|',names=svcg_column_names,low_memory=False,error_bad_lines = False)) 
    return svcg_dataset
```
